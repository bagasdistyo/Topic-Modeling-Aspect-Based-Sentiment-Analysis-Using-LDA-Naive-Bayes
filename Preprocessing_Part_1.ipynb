#Remove Number

```python
import re
def remove_numbers(text):
    text_no_numbers = re.sub(r'\d+', '', text)
    return text_no_numbers
```

```python
# Melakukan penghapusan angka dan menyimpan hasilnya di kolom baru 'after_removenumber'
dataset['after_removenumber'] = dataset['after_casefolding'].apply(remove_numbers)
```

```python
# Mengubah nilai-nilai kosong menjadi nilai null
dataset['after_removenumber'].replace('', pd.NA, inplace=True)

# Menampilkan jumlah data yang berisi nilai null atau kosong di kolom 'after_removenumber'
print("Jumlah data yang berisi nilai null atau kosong sebelum penghapusan:")
print(dataset['after_removenumber'].isnull().sum())
```

```python
# Menghapus baris yang berisi nilai null dari DataFrame
dataset.dropna(subset=['after_removenumber'], inplace=True)

# Menampilkan hasil setelah penghapusan
len(dataset)
```

#Remove Punctuation

```python
# initializing punctuations string
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~’'''

# Removing punctuations in string
# Using loop + punctuation string
dataset['after_removepunctuation'] = dataset['after_removenumber']
for ele in punc:
    dataset['after_removepunctuation'] = dataset['after_removepunctuation'].str.replace(ele, "", regex=False)
```

#Remove Special Characters

```python
import re

# Fungsi untuk menghapus karakter khusus
def remove_special_characters(text):
    # Menghapus karakter non-ASCII
    clean_text = re.sub(r'[^\x00-\x7F]+', '', text)
    return clean_text

# Mengaplikasikan fungsi pada setiap elemen dalam kolom 'after_removepunctuation'
dataset['after_removespecialcharacter'] = dataset['after_removepunctuation'].apply(lambda x: remove_special_characters(str(x)))
```

```python
# Mengubah nilai-nilai kosong menjadi nilai null
dataset['after_removespecialcharacter'].replace('', pd.NA, inplace=True)

# Menampilkan jumlah data yang berisi nilai null atau kosong di kolom 'after_removespecialcharacter'
print("Jumlah data yang berisi nilai null atau kosong sebelum penghapusan:")
print(dataset['after_removespecialcharacter'].isnull().sum())
```

```python
# Menghapus baris yang berisi nilai null dari DataFrame
dataset.dropna(subset=['after_removespecialcharacter'], inplace=True)

# Menampilkan hasil setelah penghapusan
len(dataset)
```

```python
import pandas as pd
from spellchecker import SpellChecker

# Inisialisasi SpellChecker
spell = SpellChecker(language='en')

# Daftar singkatan dan kata-kata khusus yang tidak ingin diubah
custom_words = {
    "app", "apps", "chatgpt", "gpt", "ui", "ux",
    "iot", "ai", "ml",
    "iphone", "android", "windows", "mac", "google", "microsoft",
}

# Menambahkan kata-kata khusus ke dalam kamus spellchecker
for word in custom_words:
    spell.word_frequency.add(word)

# Log untuk kata yang dikoreksi
correction_log = {}

# Fungsi untuk koreksi ejaan dengan pengecualian kata-kata khusus
def correct_spelling(text):
    # Ganti "chat gpt" dengan "chatgpt"
    text = text.replace("chat gpt", "chatgpt")

    words = text.split()
    corrected_words = []

    for word in words:
        # Cek apakah kata ada dalam daftar pengecualian
        if word in custom_words:
            corrected_words.append(word)
            continue

        correction = spell.correction(word)
        # Handle specific replacements
        if word in ["app", "apps","app's"]:
            correction = "application"
        elif word == "gpt":
            correction = "chatgpt"

        # Log koreksi jika ada perubahan
        if correction and correction != word:
            correction_log[word] = correction_log.get(word, {'correction': correction, 'count': 0})
            correction_log[word]['count'] += 1

        corrected_words.append(correction if correction else word)

    corrected_text = ' '.join(corrected_words)
    return corrected_text

# Menerapkan fungsi koreksi ejaan pada kolom 'after_removespecialcharacter'
dataset['after_spelling_correction'] = dataset['after_removespecialcharacter'].apply(correct_spelling)

# Menampilkan DataFrame setelah koreksi ejaan
print("DataFrame after spelling correction:")
print(dataset)

# Menampilkan log kata yang dikoreksi dan jumlahnya
print("\nWords corrected and their counts:")
for word, info in correction_log.items():
    print(f"{word} -> {info['correction']}: {info['count']} times")
```
